```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras.layers import Dense
```

```python
mnist = keras.datasets.mnist
```

```python
(train_features, train_labels), (test_features, test_labels) = mnist.load_data()
train_features = train_features.reshape(-1, 28, 28)
test_features = test_features.reshape(-1, 28, 28)
```

```python
from sklearn.model_selection import train_test_split

test_features, dev_features, test_labels, dev_labels = train_test_split(test_features, test_labels, test_size=0.5)

dev_features.shape, dev_labels.shape
```

# Preprocessing

```python
test_features_max = 255.0
train_features_max = 255.0
train_features = train_features.astype("float32") / train_features_max
test_features = test_features.astype("float32") / test_features_max
```

# Model Definition

```python
train_features.shape
```

```python
model = keras.Sequential()
model.add(keras.layers.Flatten())
model.add(Dense(units=128, activation="relu"))
model.add(Dense(units=128, activation="relu"))
model.add(Dense(units=10, activation="softmax"))
```

```python
model.compile(optimizer=keras.optimizers.Adam(), loss=keras.losses.SparseCategoricalCrossentropy(), metrics=["accuracy"])
```

```python
model.build(input_shape=(-1, 28, 28))
model.summary()
```

```python
hist = model.fit(train_features, train_labels, validation_data=(dev_features, dev_labels), epochs=1000, batch_size=256)
```

```python
model.evaluate(test_features, test_labels)
```
